```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("seattle-weather.csv")
```

```python
df.head()
```

```python
plt.figure(figsize = (10,6))
sns.histplot(df["weather"], bins = 20, kde = True)
plt.title("Weather")
# plt.xlable("Values")
# plt.ylable("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='weather', y='temp_min', data=df)
plt.title('Relationship between weather and min_temp')
plt.xlabel('Weather')
plt.ylabel('min temp')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='weather', data=df)
plt.title('Count of Categorical Values')
plt.xlabel('weather')
plt.ylabel('Count')
plt.show()
```

```python
df.head()
```

```python

# Convert 'diagnosis' column to numerical using Label Encoding
label_encoder = LabelEncoder()
df['date'] = label_encoder.fit_transform(df['date'])

# Convert any remaining string values to float, ignoring errors
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
